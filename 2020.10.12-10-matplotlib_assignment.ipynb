```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import interpolate
```

### Part 1

```python
x = np.arange(12)
Max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
Min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
fit_Max = np.poly1d(np.polyfit(x, Max, deg=3))
fit_Min = np.poly1d(np.polyfit(x, Min, deg=3))
```

```python
plt.scatter(x, Min, label='Min')
plt.plot(x, fit_Min(x))
plt.scatter(x, Max, label='Max')
plt.plot(x, fit_Max(x))
plt.legend(frameon=False, bbox_to_anchor=(1., .5), loc='center left')
plt.xlabel('Months')
plt.ylabel('temprature \'c')
plt.title('Temprature Trend for City X')
plt.show()
```

```python
fit_Min.coefficients
```

```python
fit_Max.coefficients
```

fitting a line with higher degree polynomials will cause over-fitting, so we're stopping at $3^\degree$

### Part 2

```python
url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic.head()
```

```python
plt.figure(figsize=(7,7))
plt.pie(titanic.sex.value_counts(), labels=['male', 'female'], autopct='%1.1f%%', colors=['orange', 'yellow']);
plt.title('Titanic Male vs Female Travellers')
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.scatterplot(x=titanic.age, y=titanic.fare, hue=titanic.sex);
plt.legend(frameon=False, loc='center left', bbox_to_anchor=(1.,0.5))
plt.title('Age to Fare per Gender', fontweight='semibold')
plt.show()
```
